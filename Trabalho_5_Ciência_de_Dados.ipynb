# **2 MODELOS DE PREVISÃO DE SÉRIES TEMPORAIS**
De acordo com Carneiro et al. (2022), os métodos de previsão de séries temporais
consideram as observações ao longo de um dado período contendo informações sobre um
padrão de comportamento. A aplicação de um modelo de previsão exige uma análise
detalhada do problema e pode ser algo complexo dependendo do número de variáveis
consideradas. Os métodos aprendem o comportamento das séries analisadas e implementam
o conhecimento adquirido para prever valores futuros, usando dados anteriores que podem
vir da própria série ou de outras séries que influenciam o comportamento.

# **2.1 MODELOS UTILIZADOS**

previsão da série temporal da quantidade de carga despachada no sistema elétrico nacional
pelas termelétricas que funcionam a óleo diesel e a óleo combustível.

# **2.1.1 MODELO SARIMA**
Os modelos estatísticos são baseados em séries históricas, geralmente combinadas
com outras metodologias autorregressiva (AR) mostrada em (BACHER; MADSEN; NIELSEN, 2009), média móvel autorregressiva (ARMA) vista em (HUANG et al., 2012), média
móvel autorregressiva integrada (ARIMA) como relata (WU; CHAN, 2012), e pode-se citar
ainda o modelo sazonal autorregressivo integrado de média móvel (SARIMA) mostrado
em (VAGROPOULOS et al., 2016).

Para a aplicação do modelo SARIMA deve-se escolher os elemntos de tendência e
de sazonalidade. Em um modelo ARIMA p é a ordem do número de defasagens de tempo,
d é a ordem de diferenciação (número de vezes que o dados tiveram valores anteriores
subtraídos) e q é a ordem do processo de média móvel de tendência (BABU; REDDY,
2014), (YANG; ZHENG; ZHANG, 2017). O Modelo SARIMA (modelo ARIMA sazonal)
é geralmente denotado como (pdq)(PDQ)S, onde P é a ordem da parte autorregressiva
sazonal (SAR), D é a ordem da diferenciação sazonal, Q é a ordem da média móvel sazonal
(SMA) e S é a duração do ciclo sazonal, ano, mês, etc (HYNDMAN; ATHANASOPOULOS,
2015).

# **2.1.2 MODELO LSTM**
As redes neurais recorrentes (RNA) se destacam com diferentes aplicações, dentre
elas pode-se citar as redes neurais recorrentes vistas em (JUNG et al., 2020) e as redes de
memória de longo-curto prazo (LSTM) mostradas em (WANG; QI; LIU, 2019).

A LSTM é capaz de lidar com a dependência de sequência, pois é um tipo de rede
neural recorrente usada em aprendizado profundo. O modelo LSTM foi proposto pela
primeira vez por Hochreiter e Schmidhuber em 1997, e mais tarde tornouse muito popular,
especialmente para abordar problemas de previsão de séries temporais (HOCHREITER;
SCHMIDHUBER, 1997). É um modelo RNN modificado que funciona bem em uma grande
variedade de problemas, e é amplamente usado atualmente. A rede neural LSTM lida
com a questão de calcular dados ao longo de um período de tempo, apresentando células
de memória no projeto da rede neural. As células de memória têm estados celulares que
armazenam experiências recentes dos dados. A cada ada momento a informação chega a
uma memória célula, o resultado é controlado através da combinação de estado da célula
e, em seguida, o estado da célula é atualizado. Agora, se houver outras informações são
recebidas pela célula de memória, a saída é processada utilizando essas novas informações e
o estado de célula atualizado. LSTM se destina a manter um problema ter dependência de
longo prazo (SUNNY; MASWOOD; ALHARBI, 2020). A rede lembra-se de informações
por um longo período de tempo, armazenando os resultados passados para serem usados
como entradas das novas interações.

# **2.1.3 MODELO NAIVE**
O modelo Naive é uma técnica de estimativa na qual os dados reais do último
período são utilizados como previsão desse período, sem nenhum ajuste ou estabelecimento
de fatores causais. É usado somente para comparação com as previsões geradas por
técnicas mais sofisticadas. Por tanto, qualquer modelo mais avançado deve apresentar
resultados melhores do que os resultados produzidos pelo Método Naive. Naive = ingênuo.
Com o intuito de realizar um comparativo para a previsão, o modelo naive será aplicado
para mostrar o quanto uma técnica de previsão mais complexa pode fornecer melhores
resultados.

Segundo Weyll, Cruz e Filho (2000), este modelo entrega como resultado da previsão
no tempo "t", o valor observado exatamente no período de "S"passos anteriores.


# **2.1.4 MODELO PROPHET**
O Prophet foi desenvolvido pela equipe de ciência de dados do Facebook em 2017
(VISHWAS; PATEL, 2020). Abiblioteca usa um modelo de série temporal qaue pode ser decomposto (ZAGDAŃSKI, 2000) com três componentes principais do modelo: tendência,
sazonalidade e feriados.

De acordo com Toharudin et al. (2020), usando o tempo como regressor, o Prophet
está tentando encaixar várias variáveis funções do tempo lineares e não lineares como
componentes. Dois modelos de tendência que cobrem os aplicativos do Facebook: um
modelo de crescimento saturado não linear e um modelo linear por partes. Um modelo
não linear é tipicamente modelado usando o modelo de crescimento logístico. Feriados
e eventos muitas vezes não seguem um padrão periódico, portanto seus efeitos não são
bem modelados por um ciclo suave. O Prophet permite que o analista forneça uma lista
personalizada de eventos passados e futuros. Uma janela em torno desses dias é considerada
separadamente e parâmetros adicionais são ajustados para modelar o efeito de feriados e
eventos.

# **2.1.5 MODELO DEEPAR**
Conforme Salinas et al. (2020), DeepAR é um método de previsão baseado em
redes neurais recorrentes autorregressivas, que aprende um modelo global a partir de
dados históricos de todas as séries temporais no conjunto de dados. Além de fornecer
uma melhor precisão de previsão do que métodos anteriores, esta abordagem tem uma
série de vantagens sobre as abordagens clássicas: (i) À medida que o modelo aprende
comportamentos sazonais e dependências em determinadas covariáveis ao longo da série
temporal, uma intervenção manual mínima no fornecimento de covariáveis é necessária para
capturar o comportamento complexo e dependente do grupo. (ii) DeepAR faz previsões
probabilísticas na forma de Amostras de Monte Carlo que podem ser usadas para calcular
estimativas de quantis consistentes para todas as subfaixas em o horizonte de previsão. (iii)
Aprendendo com itens semelhantes, este método é capaz de fornecer previsões para itens
que têm pouca ou nenhuma histórico disponível, um caso em que os métodos tradicionais
de previsão de item único falham. (vi) A abordagem não assume ruído gaussiano, mas
pode incorporar uma ampla gama de funções de verossimilhança, permitindo ao usuário
escolher uma que seja apropriada para as propriedades estatísticas dos dados.

# **3.1 METODOLOGIA APLICADA**
A análise e as previsões da série temporal da injeção de energia no sistema elétrico
pelas termelétricas a óleo combustível e a óleo diesel foram realizadas no Google Colabory.
Os dados históricos foram baixados do site da Agência Nacional de energia Elétrica
(ANEEL). Observou-se que a série temporal não é estacionária, pois apresenta um valor-p
de 0,67 e tendência e sazonalidade, além de média móvel e desvio padrão não costante.

O valor-p é definido como a probabilidade de se observar um valor da estatística
de teste maior ou igual ao encontrado. Comumente adota-se o valor de corte para rejeitar
a hipótese nula igual a 0,05, isto significa que, quando não existe nenhuma diferença,
um valor tão extremo para a estatística de teste é esperado em menos de 5% das vezes
(FERREIRA; PATINO, 2015). Nesse caso, se o valor-p for menor de 5% a hipótese nula
iria ser rejeitada (EHLERS, 2007), mas na série observada o valor-p encontrado foi 0,67.
Portanto, isto confirma que a série não é estacionária.
